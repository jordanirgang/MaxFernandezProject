## Plant based on theta
https://www.phys.lsu.edu/faculty/gonzalez/Teaching/Phys7221/DoublePendulum.pdf

Euler Lagarange
$$ \mathcal{L} = T -U $$

####notes
mass implies mass of link2, link 1 should not matter 

### Defining U
Rotational Energy
$$ U = {1 \over 2} I \omega ^2 = {1 \over 2} I {\dot \theta} ^2 $$
$$ \omega = \dot \theta $$
Linear Energy
$$ U = {1 \over 2} m v_{link_2} ^2 =  {1 \over 2} m {\dot x_{link_2}} ^2$$
$$ v = \dot x $$


### Defining T
assume shoulder joint is extended to 90 deg
$$ T =mghsin(\pi - \theta) $$

```python
import sympy as sp

mass_2,angle_I_2, length_2, height_2, time = sp.symbols('m_2 I_2 length_2 h_2 t')
mass_1,angle_I_1, length_1, height_1, time = sp.symbols('m_1 I_1 length_1 h_1 t')

theta_2 = sp.Function('theta_2')(time)
x_2 = sp.Function('x_2')(time)
theta_1 = sp.Function('theta_1')(time)
x_1 = sp.Function('x_1')(time)

U = (.5 * angle_I_1 * sp.diff(theta_1,time) **2 +# + .5 #* mass_1 * sp.diff(x_1,time)**2 
    + .5 * angle_I_2 * sp.diff(theta_2,time) **2 )# + .5 #* mass_2 * sp.diff(x_2,time)**2)
display(sp.simplify(U))
T = mass_1 * (-9.81) * height_1 * sp.sin(sp.pi/2 - theta_1) +mass_2 * (-9.81) * height_2 * sp.sin(sp.pi/2 - theta_2)
display(sp.simplify(T))
L = T- U
display(L)
```

```python

#L_theta_1_dot = L.subs(sp.diff(theta_1,time),theta_1_dot)
L_theta_1 = sp.diff(L,theta_1)
L_theta_1_dot = sp.diff(L,sp.diff(theta_1,time))
L_theta_1_dot_time = sp.diff(L_theta_1_dot,time)


#L_theta_2_dot = L.subs(sp.diff(theta_1,time),theta_1_dot)
L_theta_2 = sp.diff(L,theta_2)
L_theta_2_dot = sp.diff(L,sp.diff(theta_2,time))
L_theta_2_dot_time = sp.diff(L_theta_2_dot,time)

langrangian_theta_1 = (L_theta_1 -L_theta_1_dot_time)#.subs(theta_1_dot, sp.diff(theta_1,time))
#langrangian_x_1 = (L_x_1 -L_x_1_dot_time).subs(x_1_dot, sp.diff(x_1,time))
langrangian_theta_2 = (L_theta_2 -L_theta_2_dot_time)#.subs(theta_2_dot, sp.diff(theta_2,time))
#langrangian_x_2 = (L_x_2 -L_x_2_dot_time).subs(x_2_dot, sp.diff(x_2,time))

eqns_motion = sp.Matrix([[langrangian_theta_1,langrangian_theta_2]]).T
display(eqns_motion)
```

# small angle theory applied
$$ sin(\theta) \to \theta $$
$$ \dot \theta \to 0 $$
$$ \ddot \theta \to 0 $$

```python
eqns_motion = eqns_motion.subs(sp.sin(theta_1), theta_1)
display(eqns_motion)
eqns_motion = eqns_motion.subs(sp.diff(theta_1,time), theta_1)
display(eqns_motion)
eqns_motion = eqns_motion.subs(sp.diff(sp.diff(theta_1,time),time), theta_1)
eqns_motion = eqns_motion.subs(sp.diff(sp.diff(x_2,time),time), 0)
eqns_motion = eqns_motion.subs(sp.diff(sp.diff(x_2,time),time),0)
display(eqns_motion)

display("Theta_1 and theta2 eqns")
display(sp.expand(eqns_motion[0]))
display(sp.expand(eqns_motion[1]))
```

```python
freq = sp.symbols('s')
sp.laplace_transform(eqns_motion[0],time,freq)
```
