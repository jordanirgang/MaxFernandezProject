## Plant based on theta
Euler Lagarange
$$ \mathcal{L} = T -U $$

####notes
mass implies mass of link2, link 1 should not matter 

### Defining U
Rotational Energy
$$ U = {1 \over 2} I \omega ^2 = {1 \over 2} I {\dot \theta} ^2 $$
$$ \omega = \dot \theta $$
Linear Energy
$$ U = {1 \over 2} m v_{link_2} ^2 =  {1 \over 2} m {\dot x_{link_2}} ^2$$
$$ v = \dot x $$


### Defining T
assume shoulder joint is extended to 90 deg
$$ T =mghsin(\pi - \theta) $$

```python
import sympy as sp

mass_2,angle_I, length_2, height_2, time = sp.symbols('m_2 I_2 length_2 h_2 t')
theta = sp.Function('theta')(time)
x_2 = sp.Function('x')(time)
U = .5 * angle_I * sp.diff(theta,time) **2  + .5 * mass_2 * sp.diff(x_2,time)**2
display(sp.simplify(U))
T = mass_2 * (-9.81) * height_2 * sp.sin(sp.pi/2 - theta)
display(sp.simplify(T))
L = T- U
display(L)
```

```python
theta_dot = sp.Function('omega')
x_dot = sp.Function('v')

L_theta_dot = L.subs(sp.diff(theta,time),theta_dot)
L_theta = sp.diff(L,theta)
L_theta_dot_time = sp.diff(L_theta_dot,time)

L_x_dot = L.subs(sp.diff(x_2,time),x_dot)
L_x = sp.diff(L,x_2)
L_x_dot_time = sp.diff(L_x_dot,time)

langrangian_theta = (L_theta -L_theta_dot_time).subs(theta_dot, sp.diff(theta,time))

langrangian_x = (L_x -L_x_dot_time).subs(x_dot, sp.diff(x_2,time))


eqns_motion = sp.Matrix([[langrangian_theta,
            langrangian_x]]).T
display(eqns_motion)
```

# small angle theory applied
$$ sin(\theta) \to \theta $$
$$ \dot \theta \to 0 $$
$$ \ddot \theta \to 0 $$

```python
eqns_motion = eqns_motion.subs(sp.sin(theta), theta)
display(eqns_motion)
eqns_motion = eqns_motion.subs(sp.diff(theta,time), theta)
display(eqns_motion)
eqns_motion = eqns_motion.subs(sp.diff(sp.diff(theta,time),time), theta)
display(sp.expand(eqns_motion[0]))
```

```python
freq = sp.symbols('s')
sp.laplace_transform(eqns_motion[0],time,freq)
```
