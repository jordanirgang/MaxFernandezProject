### Defining U
Rotational Energy
$$ U = {1 \over 2} I \omega ^2 = {1 \over 2} I {\dot \theta} ^2 $$
$$ \omega = \dot \theta $$
Linear Energy- Not used unless theres a cart or something like that
$$ U = {1 \over 2} m v_{link_2} ^2 =  {1 \over 2} m {\dot x_{link_2}} ^2$$
$$ v = \dot x $$


### Defining T
assume shoulder joint is extended to 90 deg
$$ T =mghsin(\pi - \theta) $$

```python
import sympy as sp
from IPython.display import Math

mass_2,angle_I, length_2, height_2, time = sp.symbols('m_2 I_2 length_2 h_2 t')
I_bearing = sp.symbols('I_{bearing}')
theta = sp.Function('theta_1')(time)
x_2 = sp.Function('x')(time)
T = .5 * angle_I * sp.diff(theta,time) **2 # + .5 * mass_2 * sp.diff(x_2,time)**2
display(sp.simplify(U))
V = mass_2 * (-9.81) * height_2 * sp.sin(sp.pi/2 - theta)
display(sp.simplify(T))
L = T- V
display(L)
```

```python
#theta_dot = sp.diff(sp.diff(theta,time))#sp.Function('omega')
#x_dot = sp.Function('v')

L_theta_dot = sp.diff(L,sp.diff(theta,time))
display(Math(
    "{dt \mathcal{L} \over {dt( \dot {theta})}}=" + sp.latex(L_theta_dot)
))
L_theta = sp.diff(L,theta)
display(Math(
    "{dt \mathcal{L} \over {dt({theta})}} =" + sp.latex(L_theta)
))
L_theta_dot_time = sp.diff(L_theta_dot,time)
display(Math(
    "{dt{dt \mathcal{L} \over {dt( \dot {theta})}}\over dt} =" + sp.latex(L_theta_dot_time)
))

display( Math("{dt \mathcal{L} \over {dt({theta})}}-{dt{dt \mathcal{L} \over {dt( \dot {theta})}}\over dt} = 0" ))
display( Math("{dt \mathcal{L} \over {dt({theta})}}-{dt{dt \mathcal{L} \over {dt( \dot {theta})}}\over dt} = Damper" ))

langrangian_theta = L_theta -L_theta_dot_time



eqns_motion = sp.simplify(sp.Matrix([[langrangian_theta]]))

#Rayleigh dissipation function (\(R\)) and the generalized force (\(Q_{i}\))


display(sp.simplify(sp.Eq(eqns_motion[0],0)))
```

## Only looking at rolling friction with viscous lubercant inside

```python
Q  = sp.symbols('Q_{i}')
Q_torque = I_bearing * sp.diff(sp.diff(theta,time),time)
U_ball_bearing_coefs = sp.symbols('mu')
lubercant_visc = sp.symbols('v_{bear}')
frict_moment = U_ball_bearing_coefs * (lubercant_visc*sp.diff(theta))**.6
display(sp.simplify(sp.Eq(eqns_motion[0],Q)))
display(sp.simplify(sp.Eq(eqns_motion[0],sp.symbols('M_{rr}'))))

Q_torque
display(Math("differential\ eqn\ =>"+ sp.latex(sp.simplify(sp.Eq(eqns_motion[0],frict_moment)))))
```
